성별 데이터 저장하기(Type 1)

```python
import csv

import matplotlib as mpl
import matplotlib.pyplot as plt

f = open('/content/age_male_female.csv.csv', encoding = 'cp949')
data = csv.reader(f)

m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in range(0,101):
      m.append(int(row[i+3]))
      f.append(int(row[-(i+1)]))  # -1 값부터 담음 -> 100세 이상 부터 값이 담김

f.reverse() # 0세 부터 인구수 나올 수 있도록 데이터 뒤집기

print(m)
print(f)
```

성별 데이터 순서대로 저장하기(Type 2)

```python
import csv

import matplotlib as mpl
import matplotlib.pyplot as plt

f = open('/content/age_male_female.csv.csv', encoding = 'cp949')
data = csv.reader(f)

m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104]: #남성 데이터 : 3 -103번 데이터 슬라이싱 - 정수 데이터로 변화 - m 리스트에 저장
      m.append(int(i))
    for i in row[106:]: #여성 데이터 : 106번 - 끝까지 데이터 슬라이싱  - 정수 데이터로 변환 - f 리스트에 저장
      f.append(int(i))

print(m)
print(f)
```

데이터 시각화 하기

```python
import matplotlib.pyplot as plt

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
import csv

import matplotlib as mpl
import matplotlib.pyplot as plt

f = open('/content/age_male_female.csv.csv', encoding = 'cp949')
data = csv.reader(f)

m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104]: #남성 데이터 : 3 -103번 데이터 슬라이싱 - 정수 데이터로 변화 - m 리스트에 저장
      m.append(-int(i)) #남성 데이터 음수로 변환
    for i in row[106:]: #여성 데이터 : 106번 - 끝까지 데이터 슬라이싱  - 정수 데이터로 변환 - f 리스트에 저장
      f.append(int(i))

print(m)
print(f)
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.barh(range(101), m, color = 'b', label = 'Male')
plt.barh(range(101), f, color = 'r', label = 'Female')
plt.legend()
plt.show()
```
