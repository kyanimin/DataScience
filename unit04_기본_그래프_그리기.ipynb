```python
import matplotlib.pyplot as plt #라이브러리 인포트

plt.title('plotting') #타이틀 속성 추가
plt.plot([10,20,30,40], label='asc') #증가를 의미하는 asc 범례
plt.plot([40,30,20,10], label='desc') #감소를 의미하는 desc 범례

plt.legend(loc=6) #그래프에 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt #라이브러리 인포트

plt.title('color') #타이틀 속성 추가
plt.plot([10,20,30,40], color='skyblue',label='skyblue') #하늘색 그래프
plt.plot([40,30,20,10], color='pink',label='pink') #감소를 의미하는 desc 범례

plt.legend(loc=7) #그래프에 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt #라이브러리 인포트

plt.title('linestyle') #타이틀 속성 추가
plt.plot([10,20,30,40], color='r',linestyle = '--', label='dashed') #빨간색 dashed 그래프
plt.plot([40,30,20,10], color='g',ls = ':', label='dotted') #초록색 dotted 그래프 linestyle -> 표현 가능

plt.legend(loc=7) #그래프에 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt #라이브러리 인포트

plt.title('marker') #타이틀 속성 추가
plt.plot([10,20,30,40], 'r.', label='circle') #빨간색 circle 그래프
plt.plot([40,30,20,10], 'g^', label='triangle up') #초록색 triangle up 그래프 linestyle -> 표현 가능

plt.legend(loc=7) #그래프에 범례 속성 추가
plt.show()
```

데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt # 1) 라이브러리 인포트

f = open('/content/seoul.csv','r',encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)  #첫번째 행을 읽은 후 다음 행으로 이동 시키는 함수(헤더 추출)

result = [] #결측치를 제거한 최고 기온 리스트

for row in data:
  if row[-1] != '': #최고 기온이 ''(빈문자열)이 아니면,
    result .append(float(row[-1]))

result

plt.figure(dpi=200)
plt.plot(result, 'r')
plt.show()
```
