개인 데이터를 조별로 취합

1) 성별 데이터 수평 막대 그래프로 시각화(항아리 그래프) -> 본인 동네 데이터

2) 2019년 2월 연령별 인구 구조 -> 2020년 2월 연령별 인구 구조 비교(같이 그리기)

```python
import csv

import matplotlib as mpl
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
 
!apt -qq -y install fonts-nanum
 
import matplotlib.font_manager as fm

fontpath = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font = fm.FontProperties(fname=fontpath, size=9)
plt.rc('font', family='NanumBarunGothic') 
mpl.font_manager._rebuild()

f = open('/content/201903_noryangjin_pop.csv', encoding = 'cp949')
f2 = open('/content/201905_noryangjin_pop.csv', encoding = 'cp949')
data = csv.reader(f)
data2 = csv.reader(f2)

result = []
result2 = []

for row in data:

  if '노량진제1동' in row[0] :
    for i in row[3:]:
      result.append(int(i.replace(',',''))) #인구수 데이터 정수로 변환 후 result 변수에 담기

for row in data2:

  if '노량진제1동' in row[0] :
    for i in row[3:]:
      result2.append(int(i.replace(',',''))) #인구수 데이터 정수로 변환 후 result 변수에 담기


plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.style.use('ggplot') #격자 무늬 스타일로 지정; 격자 무늬 스타일로 지정 classic, dark_background
plt.plot(result, color= 'r', label = '3월')
plt.plot(result2, color= 'b', label = '5월')
plt.title("2019년 노량진제1동 지역의 인구 구조")
plt.legend()
plt.show()
```
