```python
import csv
import matplotlib.pyplot as plt # 1) 라이브러리 인포트

min_temp = 999 #최고 기온 값을 저장할 변수 -> 명확한 결과를 위해 극적인 값으로 셋팅
min_date = '' #최고 기온이 가장 높았던 날짜를 저장하는 변수

f = open('/content/jeju.csv','r',encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)  #첫번째 행을 읽은 후 다음 행으로 이동 시키는 함수(헤더 추출)


for row in data:
  
  #최고 기온 데이터가 ''(결측치) 확인
  if row[3] == '':
    row[3] = 999 #999를 넣어서 빈 문자열('')이 있던 자리라고 표시

  row[3] = float(row[3]) #기온 데이터 실수형으로 변환

  #최고 기온 정보를 저장

  if min_temp > row[3]:
    min_date = row[0]
    min_temp = row[3]

f.close()

print('기상 관측 이래 제주의 최고 기온이 가장 높았던 날은', min_date,'로', min_temp, '도 였습니다')
```
